```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Filtering netflix_df down to movies released between 1990 and 1999 inclusive
nf_movies_df = netflix_df[
    (netflix_df['type'] == 'Movie') &
    (netflix_df['release_year'] >= 1990) &
    (netflix_df['release_year'] <= 1999)
]

#Visualizing the distribution of movie durations in the 90s
plt.hist(nf_movies_df['duration'], bins = 10)
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
#Most frequent duration, in minutes, derived from the histogram
duration = 100 

#Filtering down to action movies released in the 90s
nf_movies_subset = nf_movies_df[nf_movies_df['genre'] == 'Action'] 

#Initializing the count
short_movie_count = 0 

#Iterating through the df to obtain the total count of short action movies in the 90s
for label, row in nf_movies_subset.iterrows() :
    if row['duration'] < 90 :
        short_movie_count = short_movie_count + 1
    else: short_movie_count = short_movie_count

#Printing the results of the for loop
print(short_movie_count)
```
